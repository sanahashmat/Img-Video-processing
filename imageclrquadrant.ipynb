```python
#read same image make 1st quadrant R, 2nd G, 3rd B, 4th quad as it is

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image
img = cv2.imread("/content/flower.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)   # Convert to RGB for display

h, w, c = img.shape

# Find midpoints
mid_h = h // 2
mid_w = w // 2

# Output copy
output = img.copy()

# ----- 1st Quadrant → RED -----
output[0:mid_h, 0:mid_w] = [255, 0, 0]

# ----- 2nd Quadrant → GREEN -----
output[0:mid_h, mid_w:w] = [0, 255, 0]

# ----- 3rd Quadrant → BLUE -----
output[mid_h:h, 0:mid_w] = [0, 0, 255]

# ----- 4th Quadrant → ORIGINAL -----
# (do nothing)

# ---- Display images ----
plt.figure(figsize=(10, 6))

plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(img)
plt.axis("off")

plt.subplot(1, 2, 2)
plt.title("Quadrant Output")
plt.imshow(output)
plt.axis("off")

plt.show()
```
